```python
#import the needed packages
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

#read the data
df = pd.read_csv('SA_Aqar.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# get the column
df.columns
```

```python
#rename columns
df.rename(columns={'frontyard':'front_yard'}, inplace=True)
df.columns
```

```python
#check how many missing values
df.isnull().sum()
```

```python
df['details'].mode() # I preferred not to use the result and delete the missing rows
```

```python
#Delete rows that contain missing data
da=df.dropna()
da.info()
```

```python
da['pool'] = da['pool'].replace({1: 'Yes', 0: 'No'})
da['ac'] = da['ac'].replace({1: 'Yes',0: 'No' })
da['elevator'] = da['elevator'].replace({1: 'Yes', 0: 'No'})
da['city'] = da['city'].str.strip() #delete spaces
da['city'] = da['city'].replace({'الرياض':'Riyadh',
'جدة':'Jeddah',
'الدمام':'Dammam',
'الخبر':'Khobar'})
da['city'].unique()
```

```python
#check the pool column values
da['pool'].unique()
```

```python
# Check for duplicate rows
da.duplicated()
```

```python
# It is expected that many duplicates will appear due to the nature of the dataset
# and the possibility of duplicate data,
# for example, in the price and front columns, *so I will not delete the duplicate*.
print(da['price'].value_counts())
print(da['front'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(x='city', y='price', data=da)
plt.xlabel('city')
plt.ylabel('price')
plt.title('Barplot for City')
plt.show()
```
